```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Bibliothèques chargées ✅")


```

```python
df = pd.read_csv('../data/WA_Fn-UseC_-Telco-Customer-Churn.csv')
print(f"Dimensions : {df.shape}")
df.head()
```

```python
print(df.info())
print("\n--- Valeurs manquantes ---")
print(df.isnull().sum())
```

```python
# Conversion en nombre (les espaces deviennent NaN)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Vérification
print(f"Valeurs manquantes dans TotalCharges : {df['TotalCharges'].isnull().sum()}")
```

```python
df = df.dropna(subset=['TotalCharges'])
print(f"Dimensions après nettoyage : {df.shape}")
# Tu dois avoir 7032 lignes — 11 lignes supprimées
```

```python
# Transformer Yes/No en 1/0
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
print(df['Churn'].value_counts())
```

```python
# customerID n'apporte aucune information prédictive
df = df.drop(columns=['customerID'])
print("Colonne customerID supprimée ✅")
```

```python
df.to_csv('../data/churn_clean.csv', index=False)
print("Fichier sauvegardé dans data/churn_clean.csv ✅")
```
